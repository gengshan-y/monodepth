```python
h_shift = 0
w_shift = 200
for i,l in enumerate(lines):
    t= cv2.imread('%s/%s'%(base,l.split()[0]))
    h,w,_=t.shape
    z = np.zeros((h,w,3),dtype=np.uint8)
    z[max(0,h_shift):min(h,h+h_shift),max(0,w_shift):min(w,w+w_shift),:] = t[max(0,-h_shift):min(h-h_shift,h),max(0,-w_shift):min(w-w_shift,w),:]
    
    # z[max(0,h_shift):min(h,h+h_shift),max(0,w_shift):min(w,w+w_shift),:] = t[max(0,-h_shift):min(h-h_shift,h),max(0,w_shift):min(w,w+w_shift),:]
    # z=cv2.getRectSubPix(t,(w,h),(w/2,h/2+h_shift))
    cv2.imwrite('%s/eigen-test/image_02/%04d.jpg'%(base,i),z)
    
    t= cv2.imread('%s/%s'%(base,l.split()[1]))
    h,w,_=t.shape
    z = np.zeros((h,w,3),dtype=np.uint8)
    z[max(0,h_shift):min(h,h+h_shift),max(0,w_shift):min(w,w+w_shift),:] = t[max(0,-h_shift):min(h-h_shift,h),max(0,-w_shift):min(w-w_shift,w),:]
    # z=cv2.getRectSubPix(t,(w,h),(w/2,h/2+h_shift))
    cv2.imwrite('%s/eigen-test/image_03/%04d.jpg'%(base,i),z)
```
