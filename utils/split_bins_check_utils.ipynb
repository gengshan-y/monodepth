```python
from matplotlib  import pyplot as plt
%matplotlib inline
```

```python
keys=['abs_rel', 'sq_rel','rms','log_rms','d1_all','a1','a2','a3']
```

```python
def read_res(path='log2'):
    with open(path) as f:
        lines = f.readlines()
    maps = {k:[] for k in keys}
    for l in lines:
        data = l.split(',')
        if len(data) > 1:
            for i,v in enumerate(l.split(',')):
                maps[keys[i]].append(float(v))
    return maps
```

```python
map_depth_lin = read_res('log2')
map_depth_loge = read_res('log')
map_disp_loge = read_res('log_disp_loge')
map_disp_lin = read_res('log_disp_lin')
for k,v in map_depth_loge.items():
    plt.figure()
    plt.plot(range(10,101,10),v,'r',label='linear_depth')
    plt.plot(range(10,101,10),map_depth_lin[k],'b',label='log_e_depth')
    plt.plot(range(10,101,10),map_disp_lin[k],'g',label='linear_disp')
    plt.plot(range(10,101,10),map_disp_loge[k],'k',label='log_e_disp')
    plt.ylabel(k)
    plt.legend()
```
